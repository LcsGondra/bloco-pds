```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
h = 0.01
ts = 2
T = np.arange(0, ts+h, h)

A = 3 #AMPLITUDE
f = 2 #FREQUENCIA

sinx = A*np.sin(2*np.pi*f*T)

plt.figure(figsize=(15, 15))

plt.plot(T,sinx)
plt.xlabel('t[s]')
plt.ylabel('xs(t)')
plt.title('1.1')
plt.grid()
plt.legend()
plt.show()
```

```python
A = 3 #AMPLITUDE
f = 2 #FREQUENCIA
ts = 2 #periodo
fa = 8 #frequencia de amostragem

T_fa = np.arange(0, ts+(1/fa), 1/fa)

sinx_amostrado = A*np.sin(2*np.pi*f*T_fa)

# plt.figure(figsize=(15, 15))

plt.plot(T_fa,sinx_amostrado)
plt.xlabel('t[s]')
plt.ylabel('xs(t)')
plt.title('1.2')
plt.grid()
plt.legend()

plt.show()
```
