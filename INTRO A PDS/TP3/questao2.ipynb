Gerar dois vetores com 10.000 elementos cada:
    X1: Distribuição normal com média = 1 e desvio padrão = 0.1.

```python
import random

import numpy as np

n = 10000
x1 = np.zeros((n))
for i in range(n):
    x1[i] = random.normalvariate(1, 0.1)


print(len(x1))
print(x1)
```

X2: Distribuição uniforme no intervalo [-1,1].

```python
n = 10000
x2 = np.zeros((n))
for i in range(n):
    x2[i] = random.uniform(-1, 1)


print(len(x2))
print(x2)
```

```python
print("media x1", x1.mean())
print("desvio padrao x1", x1.std())

print("media x2", x2.mean())
print("desvio padrao x2", x2.std())
```

```python
import matplotlib.pyplot as plt

T = np.arange(0, 10000, 1)

plt.subplot(211)
plt.plot(T,x1)
plt.title('normal')
plt.grid()
plt.legend()

plt.subplot(212)
plt.plot(T,x2)
plt.title('uniforme')
plt.grid()
plt.legend()


plt.show()
```
